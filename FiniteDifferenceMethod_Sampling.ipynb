```python
import numpy as np
from scipy.sparse import csr_matrix
from scipy.sparse.linalg import spsolve
import seaborn as sns
import matplotlib.pyplot as plt
from HeatFlowTomakGenerator_4 import Sample_Dataset
```

```python
num_i_grid_points = 80
num_j_grid_points = 80
Final_Time = 0.5
Num_Timesteps = 200
num_grid_points = num_i_grid_points * num_j_grid_points
```

```python
index = 0
Max_Iterations = 10
Temperature_Grids = []
Fossil_Record = []
while index <= Max_Iterations-1:
    print("Iteration",index)
    Q_11 = np.random.uniform(low = 1, high = 1000, size = (num_grid_points,))
    Q_12 = np.random.uniform(low = 1, high = 1000, size = (num_grid_points,))
    Q_22 = np.random.uniform(low = 1, high = 1000, size = (num_grid_points,))
    choice = np.random.choice([0,1,2, 3])
    print(choice)
    if choice == 0:
        Resvoir = np.random.uniform(low = 1, high = 100, size = (num_grid_points,)) * np.ones(shape = (num_grid_points,))
        Source = np.random.uniform(low = np.max(Resvoir*100), high = 100000, size = (1,)) * np.ones(shape = (num_grid_points,))
    elif choice == 1:
        Resvoir = np.random.uniform(low = 1, high = 100, size = (num_grid_points,)) * np.ones(shape = (num_grid_points,))
        Source_Magnitude = np.random.uniform(low = np.max(Resvoir*100), high = 100000, size = (1,)) 
        Source = Source_Magnitude * np.ones(shape = (num_grid_points,)) + (Source_Magnitude/100)*np.array([np.sin(x) for x in np.linspace(0,2*np.pi, num_grid_points)])
    elif choice == 2:
        Resvoir = np.random.uniform(low = 1, high = 100, size = (num_grid_points,)) * np.ones(shape = (num_grid_points,))
        Source_Magnitude = np.random.uniform(low = np.max(Resvoir*100), high = 100000, size = (1,)) 
        Source = Source_Magnitude * np.ones(shape = (num_grid_points,)) + (Source_Magnitude/100)*np.array([np.sqrt(x) for x in np.linspace(0,2*np.pi, num_grid_points)])
    elif choice == 3:
        Resvoir = np.random.uniform(low = 1, high = 100, size = (num_grid_points,)) * np.ones(shape = (num_grid_points,))
        Source_Magnitude = np.random.uniform(low = np.max(Resvoir*100), high = 100000, size = (1,)) 
        Source = Source_Magnitude * np.ones(shape = (num_grid_points,)) + (Source_Magnitude/100)*np.array([x**2 for x in np.linspace(0,2*np.pi, num_grid_points)])
    elif choice == 4:
        Resvoir = np.random.uniform(low = 1, high = 100, size = (num_grid_points,)) * np.ones(shape = (num_grid_points,))
        Source_Magnitude = np.random.uniform(low = np.max(Resvoir*100), high = 100000, size = (1,)) 
        Source = Source_Magnitude * np.ones(shape = (num_grid_points,)) + (Source_Magnitude/10)*np.array([x*np.cos(x) for x in np.linspace(0,2*np.pi, num_grid_points)])
    Temperature_Grids.append(Sample_Dataset(Q_11, Q_12, Q_22, num_i_grid_points, num_j_grid_points, Final_Time, Num_Timesteps, Resvoir, Source))
    Fossil_Record.append([Q_11, Q_12, Q_22, Source, Resvoir])
    index+=1
Temperature_Grids = np.array(Temperature_Grids)
Fossil_Record = np.array(Fossil_Record)
Temperature_Grids.shape, Fossil_Record.shape
```

```python
Q_11 = 10 * np.ones(shape = (num_grid_points,))
Q_22 = 10 * np.ones(shape = (num_grid_points,))
Q_12 = 10 * np.ones(shape = (num_grid_points,))
Resvoir = 1 * np.ones(shape  = (num_grid_points,))
Source  = 1e5 * np.ones(shape = (num_grid_points,)) + 1e4*np.array([np.sqrt(x) for x in np.linspace(0,2*np.pi, num_grid_points)])
plt.plot(Source)
```
